```python
import statsmodels.api as sm
```

```python
X=veri[["Yılda ortalama yağış (mm)"]]
```

```python
y=veri[["Temel Su Stresi"]]
```

```python
lm=sm.OLS(y,X)
```

```python
model=lm.fit()
```

```python
model.summary()
```

```python
X=sm.add_constant(X)
```

```python
lm=sm.OLS(y,X)
```

```python
model=lm.fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
model.mse_model
```
